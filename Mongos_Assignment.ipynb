```python
#Write a MongoDB query to display average number of awards won by a movie (use aggregate function with $avg operator).
for x in db.movies.aggregate([{"$group": {"_id" : None, "average" : {"$avg" : "$awards.wins"}}}]):
    print(x["average"])
```

```python
for x in db.movies.aggregate([{"$group" : {"_id" : None, "most_wins" : {"$max" : "$awards.wins"}}}]):
    print(x["most_wins"])
```
